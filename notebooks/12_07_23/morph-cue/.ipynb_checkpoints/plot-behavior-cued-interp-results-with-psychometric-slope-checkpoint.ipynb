### Load data

```python
# Specify the date you want to compare against
target_date = np.datetime64('2023-12-01')
```

```python
behav_data = {}
for bird in bird_folders:
    
    if bird.is_dir() == False:
        continue
    print(bird)
    behavior_csvs = list(bird.glob("*.csv"))
    behav_data[bird.stem] = pd.concat(
        [pd.read_csv(behavior_csv) for behavior_csv in behavior_csvs]
    )
    behav_data[bird.stem] = behav_data[bird.stem][
        behav_data[bird.stem].type_ == "normal"
    ]
    behav_data[bird.stem]['time'] = pd.to_datetime(behav_data[bird.stem]['time'])

    behav_data[bird.stem] = behav_data[bird.stem].sort_values(by='time')
    behav_data[bird.stem]['correct_bool'] = behav_data[bird.stem].correct == True
    
    # only get most recent data
    behav_data[bird.stem] = behav_data[bird.stem][behav_data[bird.stem].time.values > target_date]
    
    print(len(behav_data[bird.stem]))
    #behav_data[bird.stem] = behav_data[bird.stem][10000:]
    #print(len(behav_data[bird.stem]))
    
    # bin
    bin_edges = np.linspace(0, 127, 128)

    behav_data[bird.stem]['pos_bin'] = np.digitize(behav_data[bird.stem].interpolation_point.values, bin_edges)
    
    behav_data[bird.stem] = behav_data[bird.stem][behav_data[bird.stem].cueing == True]
    behav_data[bird.stem] = behav_data[bird.stem][behav_data[bird.stem].response != "none"]

    behav_data[bird.stem]["response_bool"] = behav_data[bird.stem].response == "left"
    behav_data[bird.stem]["interp"] = [i + j for i, j in zip(behav_data[bird.stem].left_stim.values, behav_data[bird.stem].right_stim.values)]
```

### Is the decision boundary sharper in the cued condition
- for each bird, for each interpolation, split data into cued and uncued
- compute accuracy
- compute psychometric

```python
for bird in birds:
    behav_data[bird]['bird'] = bird
behav_data_all = pd.concat([behav_data[bird] for bird in birds])
behav_data_all = behav_data_all[behav_data_all.type_ == 'normal']
behav_data_all = behav_data_all[['bird', 'interp', 'pos_bin', 'cue_id', 'response', 'correct', 'rt']]
```

```python
behav_data_all['cue_condition'] = None
behav_data_all['cue_probability'] = None
for morph, cue in cue_map.items():
    behav_data_all.loc[(behav_data_all.cue_id == cue) & (behav_data_all.interp == morph), 'cue_condition'] = 'high_prob'
    behav_data_all.loc[(behav_data_all.cue_id == cue) & (behav_data_all.interp != morph), 'cue_condition'] =  'low_prob'
    behav_data_all.loc[(behav_data_all.cue_id == cue) & (behav_data_all.interp == morph), 'cue_probability'] = 0.8
    behav_data_all.loc[(behav_data_all.cue_id == cue) & (behav_data_all.interp != morph), 'cue_probability'] = 0.1
behav_data_all.loc[(behav_data_all.cue_id == 'NC'), 'cue_condition'] = 'uncued'
behav_data_all.loc[(behav_data_all.cue_id == 'NC'), 'cue_probability'] = 0.1 / 3
behav_data_all.loc[(behav_data_all.cue_id == 'CN'), 'cue_condition'] = 'middle_prob'
behav_data_all.loc[(behav_data_all.cue_id == 'CN'), 'cue_probability'] = 1 / 3
```

### Is the psychometric slope sharpened by expectation

```python
### fit a psychometric where we have the same number of points for each condition
# .  i.e. sample 1000 rows
#   get the z score of the slope, relative to the bird/morph across samples

n_points_to_sample = 1000
n_permutations = 1000

fit_df = pd.DataFrame(
    columns=[
        "bird",
        "cue",
        "interp",
        "n_points",
        "_min",
        "_max",
        "_inflection",
        "_slope",
        #"model",
        "r_squared",
        "inflection_stderr",
    ]
)

for bird in tqdm(behav_data_all.bird.unique(), desc='bird'):
    for interp in tqdm(behav_data_all.interp.unique(), desc='interp', leave=False):
        interp_df = behav_data_all[
            (behav_data_all.bird == bird) & (behav_data_all.interp == interp)
        ]
        for cue_prob in tqdm(interp_df.cue_condition.unique(), desc='cue_prob', leave=False):
            cue_prob_df = interp_df[interp_df.cue_condition == cue_prob]
            interp_points = cue_prob_df.pos_bin.values
            response_points = np.array(cue_prob_df.response == "left").astype("int")

            for permutation in tqdm(range(n_permutations), leave=False):
                sample_idx = np.random.choice(
                    range(len(interp_points)), size=n_points_to_sample, replace=True
                )

                # fit model
                (
                    (_min, _max, _inflection, _slope),
                    results_logistic,
                    y_model,
                    r_squared,
                ) = fit_FourParameterLogistic(
                    interp_points[sample_idx],
                    response_points[sample_idx],
                    method=["leastsq"],
                )

                fit_df.loc[len(fit_df)] = [
                        bird,
                        cue_prob,
                        interp,
                        len(cue_prob_df),
                        _min,
                        _max,
                        _inflection,
                        _slope,
                        #results_logistic,
                        r_squared,
                        results_logistic.params['slope'].stderr
                    ]

# zscore within morph, to get difference between conditions
def zscore(x):
    return (x-np.mean(x))/np.std(x)

# remove iterations where fit did not converge (slope is the maximum value)
m = fit_df._slope < 99
fit_df_well_fit = fit_df[m]

for bird in fit_df_well_fit.bird.unique():
    for interp in fit_df_well_fit.interp.unique():
        m = (fit_df_well_fit.bird == bird) & (fit_df_well_fit.interp==interp)
        fit_df_well_fit.loc[m, '_slope_z'] = zscore(fit_df_well_fit[m]._slope.values)
        
```

```python
# Calculate the observed test statistic
observed_diff = df.groupby("cue_probability")["_slope"].mean().diff().iloc[-1]
```

```python
df.groupby("cue_probability")["_slope"].mean()
```

```python
from scipy.stats import pearsonr

# Calculate the observed correlation
observed_corr, _ = pearsonr(df["cue_probability"], df["_slope"])


# Function to calculate the correlation after permutation
def permutation_test(data, n_permutations=1000):
    perm_corrs = np.zeros(n_permutations)
    for i in tqdm(range(n_permutations)):
        permuted = data.copy()
        for bird in permuted["bird"].unique():
            for interp in permuted[permuted["bird"] == bird]["interp"].unique():
                mask = (permuted["bird"] == bird) & (permuted["interp"] == interp)
                permuted.loc[mask, "cue_probability"] = np.random.permutation(
                    permuted.loc[mask, "cue_probability"].values
                )
        perm_corr, _ = pearsonr(permuted["cue_probability"], permuted["_slope"])
        perm_corrs[i] = perm_corr
    return perm_corrs


# Perform the permutation test
n_permutations = 1000
perm_corrs = permutation_test(df, n_permutations)

# Calculate p-value
p_value = (np.abs(perm_corrs) >= np.abs(observed_corr)).mean()

print(f"Observed correlation: {observed_corr}")
print(f"p-value: {p_value}")
```
