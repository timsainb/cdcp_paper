```python
for offset in [False, True]:
    print('offset', offset)
    fig, axs = plt.subplots(ncols=4, figsize=(20,4))
    axi = 0
    for sr_variable in ["sr_z", "sr_z_shuffled"]:
        for passive_active in ["active", "passive"]:
            ax = axs.flatten()[axi]
            axi+=1
            ax.set_title([sr_variable, passive_active])
            if passive_active == "active":
                cue_list = ["CL1", "CL0", "NC", "CR0", "CR1"]
            else:
                cue_list = ["CL1", "NC", "CR1"]
            print(passive_active, cue_list, axi)
            mean_by_cue = (
                z_score_df[(z_score_df.passive_active == passive_active) & (z_score_df.cue.isin(cue_list))][
                    ["unit", "cue", "bird", "interp", "interp_point", sr_variable, "side"]
                ]
                .groupby(["cue", "side", "unit", "bird", "interp_point","interp"])
                .mean()
            )
            mean_by_cue = mean_by_cue.reset_index().pivot(
                columns="cue", index=["unit", "side", "bird", "interp_point", "interp"]
            )
            # subset only unit*stimuli where we have a full sampling of all cues
            mean_by_cue = mean_by_cue[np.sum(np.isnan(mean_by_cue[sr_variable].values), axis=1) == 0]
            
            # average across conditions
            mean_by_cue = (mean_by_cue.T - np.mean(mean_by_cue, axis=1).values).T
            mean_by_cue = mean_by_cue.reset_index()

            for ci, cue in enumerate(cue_list):

                x = np.arange(16)
                y = np.array(
                    [
                        np.nanmean(
                            mean_by_cue[mean_by_cue.interp_point == i][sr_variable][
                                cue
                            ].values
                        )
                        for i in x
                    ]
                )
                y_std = np.array(
                    [
                        np.nanstd(
                            mean_by_cue[mean_by_cue.interp_point == i][sr_variable][
                                cue
                            ].values
                        )
                        for i in x
                    ]
                )
                y_n = np.array(
                    [
                        np.nansum(
                            np.isnan(mean_by_cue[mean_by_cue.interp_point == i][sr_variable][
                                cue
                            ].values)==False
                        )
                        for i in x
                    ]
                )
                y_sem = y_std / np.sqrt(y_n)
                
                if offset: 
                    y = y - np.nanmean(y)

                color = cdcp_colors["cue_pal"][cue]
                ax.plot(x, y, color=color, lw=3, alpha=1)
                
    
                if False:
                    eb = ax.errorbar(
                        x,
                        y,
                        yerr=y_sem*3,
                        color=color,
                        fmt="",
                        lw=2,
                        alpha = 0.25
                    )
                if True:
                    eb = ax.fill_between(
                        x,
                        y - y_sem*3,
                        y + y_sem*3,
                        color=color,
                        #fmt="",
                        #lw=2,
                        alpha = 0.1
                    )
    for axi, ax in enumerate(axs.flatten()):
        ax.set_xticks([0, 15])
        ax.set_xticklabels([1,128], fontsize=14)
        ax.axhline(0, color="k", ls="dashed", alpha=0.25, lw=2)
        if axi == 0:
            ax.set_ylabel(
                "Spike rate rel. mean cue \n (z-scored by unit)", fontsize=16, labelpad=-0
            )
        #ax.legend(frameon=False, fontsize=14)
        #ax.set_xlim([-0.5, 1.5])
        [i.set_linewidth(5) for i in ax.spines.values()]
        
    
    if save_figs:
        ensure_dir((
            FIGURE_DIR
            / "neural_results"
            / "stimulus_spikerate"
        ))
        save_fig(
            FIGURE_DIR
            / "neural_results"
            / "cue_spikerate"
            / "cue_effect_by_interp_point_{}".format(offset),
            save_pdf=True,
            save_png=True,
        )    
    plt.show()
```

```python
def get_95_ci(x, y, n_resamples = 500, n_x_samples = 100, bounds = 95, x_range = None):
    if x_range is None:
        xmin = np.min(x)
        xmax = np.max(x)
    model_x = np.linspace(xmin, xmax, n_x_samples)
    sample_matrix = np.zeros((n_resamples, n_x_samples))
    for i in tqdm(range(n_resamples), leave=False):
        sample_index = np.random.choice(range(0, len(y)), len(y))
        X_samples = x[sample_index]
        y_samples = y[sample_index] 
        m, b = np.polyfit(X_samples, y_samples, 1)
        sample_matrix[i] = m * model_x + b
    lower_, upper_ = np.percentile(sample_matrix, ((100-bounds)/2, (100-((100-bounds)/2))), axis=0)
    return (lower_, upper_), model_x
```

```python
show_confidence_intervals = True
prob_right_dict = {
    "CL1": 0.125,
    "CL0": 0.25,
    "NC": 0.5,
    "CN": 0.5,
    "CR0": 0.75,
    "CR1": 0.875,
}
fig, axs = plt.subplots(ncols=4, figsize=(20, 4))
axi = 0
for sr_variable in ["sr_z", "sr_z_shuffled"]:
    for passive_active in ["active", "passive"]:
        ax = axs.flatten()[axi]
        axi += 1
        if passive_active == "active":
            cue_list = ["CL1", "CL0", "NC", "CR0", "CR1"]
        else:
            cue_list = ["CL1", "NC", "CR1"]
        mean_by_cue = (
            z_score_df[
                (z_score_df.passive_active == passive_active) & (z_score_df.cue.isin(cue_list))
            ][["unit", "cue", "bird", "interp", "interp_point", sr_variable, "side"]]
            .groupby(["cue", "side", "unit", "bird", "interp_point", "interp"])
            .mean()
        )
        mean_by_cue = mean_by_cue.reset_index().pivot(
            columns="cue", index=["unit", "side", "bird", "interp_point", "interp"]
        )

        # subset only unit*stimuli where we have a full sampling of all cues
        mean_by_cue = mean_by_cue[
            np.sum(np.isnan(mean_by_cue[sr_variable].values), axis=1) == 0
        ]
        mean_by_cue = mean_by_cue - np.mean(mean_by_cue, axis=0).values
        # control for differences across conditions
        mean_by_cue = (mean_by_cue.T - np.mean(mean_by_cue, axis=1).values).T
        mean_by_cue = mean_by_cue.reset_index()

        # git spikerate across cue
        sr_L = mean_by_cue[mean_by_cue.side == "L"][sr_variable][cue_list].values
        p_L = np.tile([prob_right_dict[i] for i in cue_list], len(sr_L))
        sr_L = sr_L.flatten()
        L_m, L_b = np.polyfit(p_L, sr_L, 1)
        (sr_L_lower_, sr_L_upper_), model_x = get_95_ci(p_L, sr_L)

        sr_R = mean_by_cue[mean_by_cue.side == "R"][sr_variable][cue_list].values
        p_R = np.tile([prob_right_dict[i] for i in cue_list], len(sr_R))
        sr_R = sr_R.flatten()
        R_m, R_b = np.polyfit(p_R, sr_R, 1)
        (sr_R_lower_, sr_R_upper_), model_x = get_95_ci(p_R, sr_R)

        print(scipy.stats.pearsonr(p_L, sr_L), scipy.stats.pearsonr(p_R, sr_R))

        # create a dataframe of averages
        sr_by_cue_prob_df = pd.DataFrame(columns=["cue_prob", "sr_mean", "sr_sem", "side"])
        for sr, p, side in [[sr_L, p_L, "L"], [sr_R, p_R, "R"]]:
            for cue_prob in [0.125, 0.25, 0.5, 0.75, 0.875]:
                m = p == cue_prob
                sr_mean = np.mean(sr[m])
                sr_std = np.std(sr[m])
                sr_sem = sr_std / np.sqrt(np.sum(m))
                sr_by_cue_prob_df.loc[len(sr_by_cue_prob_df)] = [
                    cue_prob,
                    sr_mean,
                    sr_sem,
                    side,
                ]

        # plot
        # fig, ax = plt.subplots(figsize=(4,4))
        for side in ["L", "R"]:
            color = cdcp_colors["reinforce"][side]["dark"]
            x = sr_by_cue_prob_df[sr_by_cue_prob_df.side == side].cue_prob.values
            y = sr_by_cue_prob_df[sr_by_cue_prob_df.side == side].sr_mean.values
            y_sem = sr_by_cue_prob_df[sr_by_cue_prob_df.side == side].sr_sem.values
            ax.scatter(x, y, color=color, label=side)
            eb = ax.errorbar(x, y, yerr=y_sem * 3, color=color, fmt="none", lw=3, alpha=0.5)
        x = np.array([0.125, 0.875])
        ax.plot(
            x,
            R_m * x + R_b,
            color=cdcp_colors["reinforce"]["R"]["dark"],
            ls="dashed",
            lw=4,
        )
        ax.fill_between(
            model_x,
            sr_R_lower_,
            sr_R_upper_,
            color=cdcp_colors["reinforce"]["R"]["dark"],
            alpha=0.25,
        )

        ax.plot(
            x,
            L_m * x + L_b,
            color=cdcp_colors["reinforce"]["L"]["dark"],
            ls="dashed",
            lw=4,
        )
        ax.fill_between(
            model_x,
            sr_L_lower_,
            sr_L_upper_,
            color=cdcp_colors["reinforce"]["L"]["dark"],
            alpha=0.25,
        )

        ax.set_xlabel("Cue Prob (P(Rr|C))", fontsize=14)
        if axi == 1:
            ax.set_ylabel("Relative spike rate \n (within unit/stimuli)", fontsize=14)
        [i.set_linewidth(5) for i in ax.spines.values()]
        #ax.legend()
        ax.set_yticks([-0.02, 0.02])
        ax.set_xlim([0, 1])
        ax.axhline(0, color="k", ls="dashed", alpha=0.25, lw=2)
    if save_figs:
        ensure_dir((
            FIGURE_DIR
            / "neural_results"
            / "stimulus_spikerate"
        ))
        save_fig(
            FIGURE_DIR
            / "neural_results"
            / "cue_spikerate"
            / "cue_effect_by_prob_class",
            save_pdf=True,
            save_png=True,
        )   
        # breakme
```
