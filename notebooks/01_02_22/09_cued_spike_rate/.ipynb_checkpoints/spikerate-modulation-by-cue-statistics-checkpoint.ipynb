```python
all_unit_statistics_df = []
for bird, timestamp, sorter in tqdm(birds, desc="birds"):

    unit_statistics_df_list = []
    for identifier in identifiers:
        unit_statistics_df = pd.read_pickle(
            DATA_DIR
            / "unit_statistics"
            / statistics_timestamp
            / identifier
            / "{}.pickle".format(bird)
        )
        unit_statistics_df_list.append(unit_statistics_df)
    print(bird, "nrows per", [len(i) for i in unit_statistics_df_list])

    # combine dfs
    unit_statistics_df = reduce(
        lambda x, y: pd.merge(x, y, on="unit", how="outer"), unit_statistics_df_list
    )
    unit_statistics_df["bird"] = bird

    all_unit_statistics_df.append(unit_statistics_df)
unit_statistics_df = pd.concat(all_unit_statistics_df)
```

### Get cuing info

```python
all_cuing_df = []
for bird, timestamp, sorter in tqdm(birds, desc="birds"):

    cuing_df = pd.read_pickle(
        DATA_DIR
        / "unit_statistics"
        / statistics_timestamp
        / "cued_trial_spikerate_z_scored_by_unit_shuffle"
        / "{}.pickle".format(bird)
    )
    print(bird, len(cuing_df))
    cuing_df["bird"] = bird

    all_cuing_df.append(cuing_df)
    
z_score_df  = pd.concat(all_cuing_df)
```

### spikerate statistic

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
z_score_df['stimulus_id'] = [str(i)+"_"+str(j) for i, j in zip(z_score_df['interp'].values, z_score_df['interp_point'].values)]
```

```python
z_score_df[:3]
```

```python
prob_right_dict = {
    'CL1': 0.125,
    'CL0': 0.25,
    'NC': 0.5,
    'CN': 0.5,
    'CR0': 0.75,
    'CR1': 0.875,
}

z_score_df.cue.values

z_score_df["cue_p_right"] = [*map(prob_right_dict.get, z_score_df.cue.values)]
```

```python
unit_stimulus = ['_'.join([str(i), str(j)]) for i, j in zip(z_score_df.stimulus_id, bird_unit)]
z_score_df['unit_stimulus'] = unit_stimulus
```

```python
z_score_df_for_r = z_score_df[z_score_df.passive_active == "active"][
    ["sr_z", "unit_stimulus", "cue", "cue_p_right", "side"]
]
z_score_df_for_r["unit_stimulus"] = pd.Categorical(z_score_df_for_r["unit_stimulus"])
z_score_df_for_r["cue"] = pd.Categorical(z_score_df_for_r["cue"])
z_score_df_for_r["side"] = pd.Categorical(z_score_df_for_r["side"])
z_score_df_for_r
```
